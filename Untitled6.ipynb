```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
ECL = pd.read_csv("export1.csv", parse_dates=["REPORTING_DATE"], index_col="REPORTING_DATE")
ECL.head()
```

```python
ECL.dropna(axis=1, how='any', inplace=True)
```

```python
ECL["PARTNER_ID"] = ECL["PARTNER_ID"].astype(str)
```

```python
ECL.loc["2020-03-31"]
```

```python
fig, ax = plt.subplots()
ax.plot(ECL.index, ECL["EAD1Y"], color="red")
ax.set_ylabel("EAD", color="red")
ax.set_xticklabels(ECL.index, rotation=90)
ax.tick_params("y", colors="red")
ax2 = ax.twinx()
ax2.plot(ECL.index, ECL["EXPECTED_LOSS_1Y"], color="blue")
ax2.set_xlabel("Time")
ax2.set_ylabel("Expected Loss", color="blue")
ax2.tick_params("y", colors="blue")
ax2.annotate("Covid Impact", xy=(pd.Timestamp("2020-03-31"), 1000000), xytext=(pd.Timestamp("2020-01"),800000),
            arrowprops={})
```
